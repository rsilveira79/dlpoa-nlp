```python
import nltk
from nltk import word_tokenize
from nltk.util import ngrams
from collections import Counter
from pprint import pprint
```

```python
corpus = "Artificial intelligence is the future. Artificial intelligence is science fiction.\
Artificial intelligence is already part of our everyday lives. \
All those statements are true, it just depends on what flavor of AI you are referring to."
```

```python
token = nltk.word_tokenize(corpus.lower())
token[1:10]
```

```python
bigrams = ngrams(token,2)
pprint(Counter(bigrams))
```

```python
trigrams = ngrams(token,3)
pprint(Counter(trigrams))
```

```python
fourgrams = ngrams(token,4)
pprint(Counter(fourgrams))
```

## Having fund with Shakespeare texts

```python
#!wget http://cs.stanford.edu/people/karpathy/char-rnn/shakespeare_input.txt
```

```python
data = open("shakespeare_input.txt").read()
```

```python
tokenizer = nltk.RegexpTokenizer(r'\w+')
shake = tokenizer.tokenize(data.lower())
shake
```

```python
shake_bigram = ngrams(shake,2, pad_left=True, pad_right=True, left_pad_symbol='<s>', right_pad_symbol='</s>')
pprint(Counter(shake_bigram).most_common(n=10))
```

```python
shake_trigram = ngrams(shake,3, pad_left=True, pad_right=True, left_pad_symbol='<s>', right_pad_symbol='</s>')
pprint(Counter(shake_trigram).most_common(n=10))
```

```python
shake_fourgram = ngrams(shake,4, pad_left=True, pad_right=True, left_pad_symbol='<s>', right_pad_symbol='</s>')
pprint(Counter(shake_fourgram).most_common(n=10))
```

```python
shake_fivegram = ngrams(shake,5, pad_left=True, pad_right=True, left_pad_symbol='<s>', right_pad_symbol='</s>')
pprint(Counter(shake_fivegram).most_common(n=10))
```

```python
freq_shake = nltk.FreqDist(shake)
freq_shake.most_common(10)
```

## Fun with Conditional Frequency Distributions

```python
cfreq_2gram = nltk.ConditionalFreqDist(nltk.bigrams(shake))
```

```python
cfreq_2gram["my"]
```

```python
cfreq_2gram["my"].most_common(10)
```

## Fun with Conditional Probability Distributions

```python
cprob_2gram = nltk.ConditionalProbDist(cfreq_2gram, nltk.MLEProbDist)
```

```python
list(cprob_2gram["my"].samples())[0:20]
```

```python
cprob_2gram["my"].prob("god")
```

```python
cprob_2gram["my"].prob("own")
```

```python
cprob_2gram["my"].prob("lord")
```

```python
cprob_2gram["my"].prob("son")
```

```python
cfreq_2gram["what"].most_common(10)
```

## Making predictions - Phrase generator

```python
seed = 'magic'
#seed = 'adriano'
#seed = 'what'

first = cprob_2gram[seed].max()
first
```

```python
second = cprob_2gram[first].max()
second
```

```python
third = cprob_2gram[second].max()
third
```

```python
fourth = cprob_2gram[third].max()
fourth
```

```python
fifth = cprob_2gram[fourth].max()
fifth
```

```python
final_phrase = seed + " " + first + " " + second + " " + third + " " + fourth + " " + fifth + "."
final_phrase
```
